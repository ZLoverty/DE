# Notebook for plotting figures

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 1 Size control

```python
data = pd.read_csv('../Data/DE Size Control - Data.csv')
```

```python
data
```

### 1.1 Fix the ratio between outer and inner sum

Plot the inner and outer droplet sizes as a function of the ratio between middle and inner flow rates ($Q_m/Q_i$).

```python
data1 = data.loc[data['Outer flow rate']==1000]
qmqi = data1['Middle flow rate'] / data1['Inner flow rate']
plt.figure(figsize=(5, 3), dpi=150)
plt.errorbar(qmqi, data1['Outer droplet radius'], data1['Outer droplet radius error'],
            ls='', marker='o', label='$R$', markersize=15, mfc=(0,0,0,0), mec='black', color='black')
plt.errorbar(qmqi, data1['Inner droplet radius'], data1['Inner droplet radius error'],
            ls='', marker='o', label='$r$', markersize=8, color='black')
plt.xlabel('$Q_m/Q_i$')
plt.ylabel('Droplet radii (px)')
plt.legend()
```

### 1.2 Fix the ratio between outer sum and inner
Plot the inner and outer droplet sizes as a function of the ratio between outer and middle flow rates ($Q_o/Q_m$).

```python
data2 = data.loc[data['Sum outer middle']==1700]
qmqi = data2['Outer flow rate'] / data2['Middle flow rate']
plt.figure(figsize=(5, 3), dpi=150)
plt.errorbar(qmqi, data2['Outer droplet radius'], data2['Outer droplet radius error'],
            ls='', marker='o', label='$R$', markersize=15, mfc=(0,0,0,0), mec='black', color='black')
plt.errorbar(qmqi, data2['Inner droplet radius'], data2['Inner droplet radius error'],
            ls='', marker='o', label='$r$', markersize=8, color='black')
plt.xlabel('$Q_o/Q_m$')
plt.ylabel('Droplet radii (px)')
plt.legend()
```
