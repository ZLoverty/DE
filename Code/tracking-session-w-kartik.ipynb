```python
frame_list = []
for label in traj.Label:
    frame = int(label.split("/")[0].split(":")[2])
    frame_list.append(frame)
```

```python
len(frame_list)
```

```python
t1 = traj.assign(frame=frame_list)
```

```python
t1
```

```python
0.1 * 30
```

```python
t2 = tp.link(t1, 10, pos_columns=["X", "Y"])
```

```python
t3 = tp.filter_stubs(t2, 50)
```

```python
t3
```

```python
# particle number histogram
count_list = []
for _, g in t3.groupby("particle"):
    count_list.append(len(g))
plt.bar(t3.particle.drop_duplicates().astype("string"), count_list)
plt.xticks(t3.particle.drop_duplicates().astype("string"))
```

```python
tp.plot_traj(t3, pos_columns=["X", "Y"], label="particle")
```
