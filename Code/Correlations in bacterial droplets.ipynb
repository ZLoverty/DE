# Correlations in bacterial droplets

In this notebook, I visualize the correlation analysis of bacterial motion in water droplets.

## 1 Correlation time

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
log = pd.read_excel(io="../Data/structured_log.ods", sheet_name="main")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(log["Droplet size"], log["tau"], c=log["Initial mean velocity (10 s)"])
plt.xlabel("Droplet diameter (um)")
plt.ylabel("Correlation time (s)")
plt.xlim([0, 140])
plt.ylim([0, 1])
plt.colorbar(label="mean velocity (um/s)")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(log["Droplet size"], log["tau"]*log["Initial mean velocity (10 s)"], c=log["Initial mean velocity (10 s)"])
plt.xlabel("Droplet diameter (um)")
plt.ylabel("$v\\tau$ (um)")
plt.xlim([0, 140])
plt.ylim([0, 5])
plt.colorbar(label="mean velocity (um/s)")
```
